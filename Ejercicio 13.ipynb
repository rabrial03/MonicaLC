```python
!pip install ed25519

import ed25519
import hashlib

# Cargar la clave privada
with open("/ed25519-priv", "rb") as f:
    private_key = f.read()


signing_key = ed25519.SigningKey(private_key)
public_key = signing_key.get_verifying_key()
# Mensaje a firmar
mensaje = bytes("Firmamos con la curva 25519", "utf8")
hash = hashlib.sha256()
hash.update(mensaje)
digest = hash.digest()
signature = signing_key.sign(digest)
# Convertir la firma a Hexadecimal
signature_hex = signature.hex()
# Imprimir la firma en Hexadecimal
print("Firma en hexadecimal:", signature_hex)
```
