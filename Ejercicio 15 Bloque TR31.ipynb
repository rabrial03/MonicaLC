```python
!pip install Pycryptodome
from Crypto.Cipher import AES
import binascii

# Bloque TR31
tr31_block = "D0144D0AB00S000042766B9265B2DF93AE6E29B58135B77A2F616C8D515ACDBE6A5626F79FA7B4071E9EE1423C6D7970FA2B965D18B23922B5B2E5657495E03CD857FD37018E111B"

# Clave de transporte para desencriptar el bloque (AES-128)
key_transport = "A1A10101010101010101010101010102"

# Convierto clave de transporte y bloque TR31 a bytes
key_transport_bytes = binascii.unhexlify(key_transport)
encrypted_block_bytes = binascii.unhexlify(tr31_block[16:])  # Omitir encabezado TR31 (16 caracteres iniciales)

# Creo objeto AES en modo ECB
cipher = AES.new(key_transport_bytes, AES.MODE_ECB)

# Desencripto el bloque
decrypted_block = cipher.decrypt(encrypted_block_bytes)

# Separo los datos desencriptados (clave y KCV)
key_bytes = decrypted_block[:-3]  # Clave desencriptada (omitimos los Ãºltimos 3 bytes)
kcv_bytes = decrypted_block[-3:]  # Key Check Value (3 bytes al final)

# Muestro los resultados
print(f"Clave desencriptada (hex): {binascii.hexlify(key_bytes).decode()}")
print(f"KCV (Key Check Value, hex): {binascii.hexlify(kcv_bytes).decode()}")
```
