```python
!pip install pyjwt
import hmac
import hashlib
import base64

# Datos necesarios
header = '{"typ":"JWT","alg":"HS256"}'
payload = '{"usuario":"Don Pepito de los palotes","rol":"isNormal","iat":1667933533}'
secret_key = "Con KeepCoding aprendemos"

# Paso 1: Codificar header y payload en Base64
header_b64 = base64.urlsafe_b64encode(header.encode()).decode().strip("=")
payload_b64 = base64.urlsafe_b64encode(payload.encode()).decode().strip("=")

# Paso 2: Crear el mensaje para firmar (header.payload)
message = f"{header_b64}.{payload_b64}"

# Paso 3: Calcular la firma HMAC con SHA-256
signature = hmac.new(secret_key.encode(), message.encode(), hashlib.sha256).digest()

# Paso 4: Codificar la firma en Base64 para que sea compatible con el formato JWT
signature_b64 = base64.urlsafe_b64encode(signature).decode().strip("=")

# Resultado final del JWT
jwt = f"{message}.{signature_b64}"

print("JWT generado:")
print(jwt)
```
